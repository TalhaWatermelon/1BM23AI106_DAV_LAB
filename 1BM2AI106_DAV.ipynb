```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv('New Microsoft Excel Worksheet (4).csv')
print(df)
```

```python
x=df['cgpa']
y=df['sgpa']
plt.xlabel('sgpa')
plt.ylabel('cgpa')
plt.plot(x,y)
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
mx=x.mean()
print(mx)
mox=x.mode()
print(mox)
my=y.mean()
print(my)
moy=y.mode()
print(moy)
```

```python
x.plot.box()
plt.title("box plot for cgpa")
plt.ylabel("cgpa")
plt.show()
```
